```python
import os

os.makedirs("ds_shashivardhan/csv_files", exist_ok=True)
os.makedirs("ds_shashivardhan/outputs", exist_ok=True)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sentiment=pd.read_csv('/content/ds_shashivardhan/csv_files/fear_greed_index.csv')
sentiment
```

```python
trader=pd.read_csv('/content/ds_shashivardhan/csv_files/historical_data.csv')
trader
```

```python
sentiment.info()
```

```python
trader.info()
```

```python
trader['Timestamp IST']=pd.to_datetime(trader['Timestamp IST'],format='%d-%m-%Y %H:%M')
trader['date']=pd.to_datetime(trader['Timestamp IST'].dt.date)
trader['date']
```

```python
sentiment['date']=pd.to_datetime(sentiment['date'])
sentiment['date']
```

```python
df=trader.merge(sentiment[['date','classification','value']],on='date',how='left')
df.head()
```

```python
df.to_csv("ds_shashivardhan/csv_files/merged_trader_sentiment.csv", index=False)
```

```python
df.size,df.shape
```

```python
df.isnull().sum()
```

```python
df=df.dropna(subset=["classification"])
```

```python
df.isnull().sum()
```

```python
df['classification'].value_counts()
```

```python
avg_pnl=df.groupby('classification')['Closed PnL'].mean()
avg_pnl
```

```python
total_pnl=df.groupby('classification')['Closed PnL'].sum()
total_pnl
```

```python
df['profit/loss']=df['Closed PnL'] > 0
len(df[df['profit/loss']==True]),len(df[df['profit/loss']==False])
```

```python
win_rate =df.groupby("classification")['profit/loss'].mean()*100
print(win_rate)
```

```python
avg_size=df.groupby('classification')['Size USD'].mean()
avg_size
```

```python
avg_fee=df.groupby('classification')['Fee'].mean()
avg_fee
```

```python
coin_perf = df.groupby(['Coin', 'classification'])['Closed PnL'].mean()
print(coin_perf)
```

```python
acc=df.groupby('Account')['Closed PnL'].sum()
top_acc=acc.sort_values(ascending=False).head(10)
print(top_acc)
```

```python
top_trader_acc=df[df['Account'].isin(top_acc.index)].groupby('classification')['Closed PnL'].mean()
top_trader_acc
```

```python
df.boxplot(column='Closed PnL', by='classification')
plt.title('PnL Distribution: Fear vs Greed')
plt.ylabel('Closed PnL')
plt.savefig('ds_shashivardhan/outputs/PnL Distribution: Fear vs Greed.png')
plt.show()
```

```python
avg_pnl.plot(kind='bar')
plt.title('Average PnL per Trade: Fear vs Greed')
plt.ylabel('Mean Closed PnL')
plt.savefig('ds_shashivardhan/outputs/Average PnL per Trade: Fear vs Greed.png')
plt.show()
```

```python
win_rate.plot(kind='bar')
plt.title('Win Rate (%) in Fear vs Greed Markets')
plt.ylabel('Win Rate (%)')
plt.savefig('ds_shashivardhan/outputs/Win Rate (%) in Fear vs Greed Markets.png')
plt.show()
```

```python
avg_size.plot(kind='bar')
plt.title('Average Trade Size (USD): Fear vs Greed')
plt.ylabel('Avg Size USD')
plt.savefig('ds_shashivardhan/outputs/Average Trade Size (USD): Fear vs Greed.png')
plt.show()
```

```python
summary = df.groupby("classification").agg(
    Avg_PnL=('Closed PnL', 'mean'),
    Total_PnL=('Closed PnL', 'sum'),
    Win_Rate=('profit/loss', 'mean'),
    Avg_Trade_Size=('Size USD', 'mean'),
    Avg_Fee=('Fee', 'mean')
)
summary['Win_Rate'] = summary['Win_Rate'] * 100
print(summary)
```

```python
summary.to_csv("ds_shashivardhan/csv_files/sentiment_summary.csv")
```
